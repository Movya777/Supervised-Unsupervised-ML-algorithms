In this notebook, I implemented dimensionality reduction using PCA


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
# Load the dataset
from sklearn.datasets import load_breast_cancer
```

```python
cancer_dataset=load_breast_cancer()
```

```python
cancer_dataset.keys()
```

```python
print(cancer_dataset.DESCR)
```

```python
# Create a dataframe
df=pd.DataFrame(cancer_dataset['data'],columns=cancer_dataset['feature_names'])
df.head()
```

### Step 1: Standardization

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
scaled_data
```

### Appplying PCA algorithm

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2) # 2 dimension
```

```python
data_pca=pca.fit_transform(scaled_data)
data_pca
```

The dataset is converted into 2 feature columns

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(8,6))
plt.scatter(data_pca[:,0],data_pca[:,1],c=cancer_dataset['target'],cmap='plasma')
plt.xlabel('First principal component')
plt.ylabel('Second principal component')
```

Hooray!!!

The distinction is clearly visible after the dimensioanlity reduction.

Learnings:

1. How to use PCA for dimensionality reduction
2. Look at the variance captured by the components
